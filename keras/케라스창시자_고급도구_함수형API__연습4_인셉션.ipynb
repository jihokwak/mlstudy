```
from keras import layers
from keras.layers import Input
from keras.models import Model
```

```
x = Input(shape=(None,None,128), dtype='float32')
```

```
branch_a = layers.Conv2D(128, 1, activation='relu', strides=2)(x)
#모든 가지는 동일한 스트라이드(2)를 사용. 출력 크기를 동일하게 만들어 하나로 합치기 위함.
branch_b = layers.Conv2D(128, 1, activation='relu')(x)
branch_b = layers.Conv2D(128, 3, activation='relu', strides=2)(branch_b)
#두 번째 합성곱 층에서 스트라이드 적용
branch_c = layers.AveragePooling2D(3, strides=2)(x)
branch_c = layers.Conv2D(128, 3, activation='relu', strides=2)(branch_c)
#평균 풀링 층에서 스트라이드 적용
branch_d = layers.Conv2D(128, 1, activation='relu')(x)
branch_d = layers.Conv2D(128, 3, activation='relu')(branch_d)
branch_d = layers.Conv2D(128, 3, activation='relu', strides=2)(branch_d)

output = layers.concatenate([branch_a, branch_b, branch_c, branch_d], axis=-1)
#모든 가지의 출력을 연결하여 모듈의 출력을 만듬.
```

```
model = Model(x, output)
```

```
model.summary()
```
